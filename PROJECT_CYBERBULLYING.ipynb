**PROSES CLEANING**

---



```python
import re
import string
import nltk

# Fungsi untuk menghapus URL
def remove_URL(tweet):
    if tweet is not None and isinstance(tweet, str):
        url = re.compile(r'https?://\S+|www\.\S+')
        return url.sub(r'', tweet)
    else:
        return tweet

# Fungsi untuk menghapus HTML
def remove_html(tweet):
    if tweet is not None and isinstance(tweet, str):
        html = re.compile(r'<.*?>')
        return html.sub(r'', tweet)
    else:
        return tweet

# Fungsi untuk menghapus emoji
def remove_emoji(tweet):
    if tweet is not None and isinstance(tweet, str):
        emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F700-\U0001F77F"  # alchemical symbols
            u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
            u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
            u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
            u"\U0001FA00-\U0001FA6F"  # Chess Symbols
            u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
            u"\U0001F004-\U0001F0CF"  # Additional emoticons
            u"\U0001F1E0-\U0001F1FF"  # flags
                               "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', tweet)
    else:
        return tweet

# Fungsi untuk menghapus simbol
def remove_symbols(tweet):
    if tweet is not None and isinstance(tweet, str):
        tweet = re.sub(r'[^a-zA-Z0-9\s]', '', tweet)  # Menghapus semua simbol
    return tweet

# Fungsi untuk menghapus angka
def remove_numbers(tweet):
    if tweet is not None and isinstance(tweet, str):
        tweet = re.sub(r'\d', '', tweet)  # Menghapus semua angka
    return tweet

def hapus_username(tweet):
    pattern = r'@[\w]+'
    tweet = re.sub(pattern, '', tweet)
    return tweet


df['cleaning'] = df['full_text'].apply(lambda x: remove_URL(x))
df['cleaning'] = df['cleaning'].apply(lambda x: hapus_username(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_html(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_emoji(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_symbols(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_numbers(x))

df.head(5)
```

**PROSES CASE FOLDING**

---



```python
def case_folding(text):
    if isinstance(text, str):
        lowercase_text = text.lower()
        return lowercase_text
    else:
        return text

df['case_folding'] = df['cleaning'].apply(case_folding)
df.head(5)
```

**TOKENIZATION**

---



```python
def tokenize(text):
    tokens = text.split()
    return tokens

df['tokenize'] = df['case_folding'].apply(tokenize)

df.head(5)
```

**PROSES STOPWORD REMOVAL**

---



```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('indonesian')
```

```python
def remove_stopwords(text):
    return [word for word in text if word not in stop_words]

df['stopword removal'] = df['tokenize'].apply(lambda x: remove_stopwords(x))

df.head(5)
```

```python
import pandas as pd

# Baca dataset
data = pd.read_csv("Hasil_Preprocessing_Data.csv")
df_dataset1 = pd.DataFrame(data[['Tanggal','Waktu','username','quote_count','reply_count','retweet_count','favorite_count','steming_data']])


# Baca kamus kata tidak baku
kamus_data = pd.read_excel("kamuskatabaku.xlsx")
kamus_tidak_baku = dict(zip(kamus_data['tidak_baku'], kamus_data['kata_baku']))
```

```python
import pandas as pd

# Fungsi penggantian kata tidak baku
def replace_taboo_words(text, kamus_tidak_baku):
    if isinstance(text, str):
        words = text.split()
        replaced_words = []
        kalimat_baku = []
        kata_diganti = []
        kata_tidak_baku_hash = []

        for word in words:
            if word in kamus_tidak_baku:
                baku_word = kamus_tidak_baku[word]
                if isinstance(baku_word, str) and all(char.isalpha() for char in baku_word):
                    replaced_words.append(baku_word)
                    kalimat_baku.append(baku_word)
                    kata_diganti.append(word)
                    kata_tidak_baku_hash.append(hash(word))
                else:
                    replaced_words.append('<karakter eksplisit>')
                    kalimat_baku.append(word)
            else:
                replaced_words.append(word)
        replaced_text = ' '.join(replaced_words)
    else:
        replaced_text = ''
        kalimat_baku = []
        kata_diganti = []
        kata_tidak_baku_hash = []

    return replaced_text, kalimat_baku, kata_diganti, kata_tidak_baku_hash
```

```python
# Terapkan fungsi penggantian kata tidak baku
df_dataset1['hasil_normalisasi'], df_dataset1['Kata_Baku'], df_dataset1['Kata_Tidak_Baku'], df_dataset1['Kata_Tidak_Baku_Hash'] = zip(*df_dataset1['steming_data'].apply(lambda x: replace_taboo_words(x, kamus_tidak_baku)))

df_dataset1.head(5)
```

```python
import pandas as pd

class SentimentClassifier:
    def __init__(self):
        self.ya = ['paling cerdas', 'paham tidak tugas walikota', 'Bank BUMN itu diluar ranah walikota',
                   'besok banget lagi kelakuan sama yang lain', 'si paling tau masalah perbank', 'sudah disuruh hapus postingan',
                   'Ceritan mau hapus histori kebobrokan inti kaulu dia bukan anak presiden', 'tidak mungkin dia jadi cawapres',
                   'kapan mundur juga berani tidak', 'pakai fasilitas negara', 'prabowo kapan mundur', 'tidak tau malu']
        self.tidak = ['tidak beres bapak', 'setiap permasalahan yang ada', 'mohon untuk diadukan kepada kepala wilayah masing masing',
                      'dari kepala wilayah tingkat terendah dulu', 'tingkat sedang', 'baru tingkat teratas begitu', 'jadi tidak makan gaji buta',
                      'sehingga tahu betul', 'tupoksi','semangat', 'kalau mau jujurbisa dilihat kekayaan sebelum',
                      'sesudah menjabat setelah tidak menjabat', 'nanti dikurangi total gaji', 'pesangon', 'serta pengeluaran per bulan berapa',
                      'bisa kelihatan njomplang', 'proyek boleh tidak pegang langsung', 'tapi bisa atas nama orang lain atau kerjasama',
                      'anjut agar rakyat solo makin sejahtera', 'prabowo gibran maju terus pantang mundur', 'ayo maju terus pantang mundur',
                      '82% rakyat Indonesia dibelakangmu', 'Sudah lihat debat capres pertama', 'sudah jelas ditekankan',
                      'Coba tonton di segmen pertanyaan anatara capres']

    def determine_sentiment(self, text):
        positive_count = sum(keyword in text.lower() for keyword in self.ya)
        negative_count = sum(keyword in text.lower() for keyword in self.tidak)

        if positive_count > negative_count:
            return 'ya'
        elif negative_count > positive_count:
            return 'tidak'
        else:
            return 'tidak'

sentiment_classifier = SentimentClassifier()

df['Sentiments'] = df['hasil_normalisasi'].apply(sentiment_classifier.determine_sentiment)
df.head(5)
```

# ***Algoritma K-Nearest Neighbor (KNN)***

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
```

```python
import pandas as pd

# Membaca file CSV dengan nama "Hasil_Labelling_Data.csv" ke dalam sebuah DataFrame menggunakan fungsi read_csv() dari pandas
data = pd.read_csv("Hasil_Labelling_Data.csv")

# Menampilkan 10 baris pertama dari DataFrame menggunakan fungsi head()
data.head(10)
```

```python
from sklearn.model_selection import train_test_split

# Bagi data menjadi data pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(data['hasil_normalisasi'], data['Sentiments'], test_size=0.2, random_state=42)
```

```python
# Simpan data latih ke file
train_set = pd.DataFrame({'text': X_train, 'Sentiments': y_train})
train_set.to_csv('train_data.csv', index=False)
```

```python
# Simpan data uji ke file
test_set = pd.DataFrame({'text': X_test, 'Sentiments': y_test})
test_set.to_csv('test_data.csv', index=False)
```

```python
# Menampilkan informasi jumlah data
print(f'Jumlah Data Latih: {len(X_train)}')
print(f'Jumlah Data Uji: {len(X_test)}')
```

```python
import matplotlib.pyplot as plt

# Jumlah data latih dan data uji
train_size = len(X_train)
test_size = len(X_test)

# Membuat plot
plt.figure(figsize=(8, 6))
bars = plt.bar(['Data Latih', 'Data Uji'], [train_size, test_size], color=['blue', 'orange'])

# Menambahkan label untuk setiap bar (dalam kurung persentase)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 20, f'{height} ({height / (train_size + test_size) * 100:.2f}%)',
             ha='center', va='bottom')

plt.title('Jumlah Data Latih dan Data Uji')
plt.xlabel('Jenis Data')
plt.ylabel('Jumlah Data')
plt.show()
```

```python
# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Ubah teks menjadi vektor TF-IDF
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# Inisialisasi KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)

# Latih classifier
knn.fit(X_train_tfidf, y_train)

# Prediksi label data uji
y_pred = knn.predict(X_test_tfidf)
```

```python
from sklearn.metrics import classification_report

# Hitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("===============")

# Hitung laporan klasifikasi
report = classification_report(y_test, y_pred)

# Tampilkan laporan klasifikasi
print(report)
```

```python
from sklearn.metrics import confusion_matrix

# Hitung matriks kebingungan
cm = confusion_matrix(y_test, y_pred)

# Cetak matriks kebingungan
print("Confusion Matrix:")
print(cm)
```

```python
import matplotlib.pyplot as plt

# Visualisasi matriks kebingungan menggunakan heatmap tanpa seaborn
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap='Blues')
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = range(len(set(y_test)))
plt.xticks(tick_marks, ['tidak', 'ya'], rotation=45)
plt.yticks(tick_marks, ['tidak', 'ya'])
plt.xlabel('Predicted')
plt.ylabel('Actual')

# Menambahkan teks pada sel heatmap
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'),
                 ha="center", va="center",
                 color="white" if cm[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
```
